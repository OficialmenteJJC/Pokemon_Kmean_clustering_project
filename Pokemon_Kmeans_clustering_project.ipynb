```python
import pandas as pd
```

```python
df = pd.read_csv('pokemon.csv')
df.head()
```

```python

important_col = ['hp','height_m', 'attack',
'base_total','defense','weight_kg',
'speed', 'name']

X = df[important_col].dropna()
X_big = X
X = X.iloc[:,:-1]
```

```python
from sklearn.cluster import KMeans
kmean= KMeans(n_clusters = 4)
kmean.fit(X)

labels = kmean.predict(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X.iloc[:,0], X.iloc[:,1],
c=labels)
plt.show()
```

```python
Ks = range(1, 10)
km = [KMeans(n_clusters=i) for i in Ks]
score = [km[i].fit(X).score(X) for i in range(len(km))]
plt.plot(Ks, score)
```

```python
import seaborn as sns
sns.pairplot(X)
```

```python
import plotly.express as px
px.scatter_matrix(X, color = [str(i) for i in labels],
hover_name = X_big['name'])
```
